```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
def circle_in_Lp(p):
  alpha = np.linspace(-np.pi, np.pi, 10000)
  r = np.power(np.power(np.abs(np.sin(alpha)), p) + np.power(np.abs(np.cos(alpha)), p), -1/p)
  return r*np.cos(alpha), r*np.sin(alpha)
```

```python
plt.figure(figsize=(8, 8))
x, y = circle_in_Lp(p = 1)
plt.fill(x, y, c = 'r')
plt.show()
```

```python
plt.figure(figsize=(8, 8))
x, y = circle_in_Lp(p = 2)
plt.fill(x, y, c = 'b')
plt.show()
```

```python
plt.figure(figsize=(8, 8))
x, y = circle_in_Lp(p = 3)
plt.fill(x, y, c = 'g')
plt.show()
```
